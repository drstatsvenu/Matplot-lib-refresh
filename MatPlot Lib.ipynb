# Boxplot

```python
# Example is taken from boxplot documentation of matplotlib

df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df.plot.box()
```

```python
color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray')
df.plot.box(color=color, sym='r+')
```

```python
# Vertical means whether a box looks vertcal or horizontal and positions where the boxes are w.r.to y axis

df.plot.box(vert = False,positions = [1,3,4,6,7])
```

```python
# You can create a stratified output using the by keyword in boxplot

q = pd.DataFrame(np.random.randn(10,2),columns = ['col1','col2'])
q['x'] = pd.Series(['A','A','A','A','A','B','B','B','B','B'])
plt.figure()
q.boxplot(by = 'x',figsize = (15,5))
```

```python
# please note that the graph two DISTINCT results for plot.box() and .boxplot() command
# you can also plot a subset of columns to the boxplot command like ['col1','col2'],by =['x','y']
```

```python
# Plotting with error bars
# Data taken from document
ix3 = pd.MultiIndex.from_arrays([['a', 'a', 'a', 'a', 'b', 'b', 'b', 'b'], ['foo', 'foo', 'bar', 'bar', 'foo', 'foo', 'bar', 'bar']], names=['letter', 'word'])
df3 = pd.DataFrame({'data1': [3, 2, 4, 3, 2, 4, 3, 2], 'data2': [6, 5, 7, 5, 4, 5, 6, 5]}, index=ix3)
gp3 = df3.groupby(level=('letter', 'word'))
means = gp3.mean()
errors = gp3.std()
```

```python
means
```

```python
errors
```

```python
fig,ax = plt.subplots()
means.plot.bar(yerr=errors,ax = ax)
# understand it better
```

## plotting tables

```python
fig,ax = plt.subplots(1,1)
df = pd.DataFrame(np.random.randn(5,3),columns = ['A','B','C'])
ax.get_xaxis().set_visible(False)
df.plot(table = True,ax =ax)

# why giving ax = ax is necessary is it only to sync two subplots
```

```python
# np.round rounds off a number to given decimals
# np.around evenly rounds off

fig,ax = plt.subplots(1,1)
ax.get_xaxis().set_visible(False)
df.plot(table=np.round(df.T, 2), ax=ax)
```

```python
# you can also specify the position of box to come
# see the following example
from pandas.plotting import table
fig,ax = plt.subplots(1,1)
table(ax,np.round(df.describe(),2),loc = 'upper right',colWidths = [0.2,0.2,0.2])
df.abs().plot(ax=ax,ylim = (0,4),legend = None,figsize = (10,10))
```
